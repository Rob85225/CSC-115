```python
import pandas as pd

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/Rob85225/CSC-115/main/nycflights.csv")

# Preview the data
df.head()
```

```python
# American Airlines (AA) and Delta Airlines (DL)
aa_count = df[df['carrier'] == 'AA'].shape[0]
dl_count = df[df['carrier'] == 'DL'].shape[0]

print(f"American Airlines flights: {aa_count}")
print(f"Delta Airlines flights: {dl_count}")
```

```python
# Filter for JFK origin
jfk_df = df[df['origin'] == 'JFK']

# Top destinations from JFK
top_dest_jfk = jfk_df['dest'].value_counts().head(5)

print("Top destinations from JFK:")
print(top_dest_jfk)
```

```python
# Group by origin and destination
origin_dest_counts = df.groupby(['origin', 'dest']).size().reset_index(name='flight_count')

print(origin_dest_counts.head())
```

```python
missing_dep_time = df['dep_time'].isna().sum()
print(f"Missing values in dep_time: {missing_dep_time}")
```

```python
# Group by carrier and calculate average departure delay
avg_dep_delay = df.groupby('carrier')['dep_delay'].mean().sort_values(ascending=False)

# Airline with highest average delay
highest_delay_airline = avg_dep_delay.idxmax()
highest_delay_value = avg_dep_delay.max()

print(f"Airline with highest average departure delay: {highest_delay_airline} ({highest_delay_value:.2f} minutes)")
```
